```python
from pathlib import Path
from PIL import Image
```

```python
p = Path("./training-runs/00003-female_tfr-paper256-kimg10000")
```

```python
g = p.glob("fakes*.png")
```

```python
def resize(img_path):
    print(img_path)
    with Image.open(img_path) as img:
        f_name = str(img_path).split("/")[-1].split(".")[0]
        print(f_name)
        img2 = img.crop((0, 0, 4096, 4096)) 
        img2.save(f'./out/fakes/{f_name}.jpg')
```

```python
for i in g:
    resize(i)
```

```python
p_resized = Path("./out/fakes/")
```

```python
g = p_resized.glob("fakes*.jpg")
```

```python
l =list(g)
```

```python
l.sort()
```

```python
l[0]
```

```python

# Generate video.
import moviepy.editor
from moviepy.editor import *
```

```python
clips = [ImageClip(str(i)).set_duration(0.5) for i in l]
```

```python
mp4_file = f'out/train.mp4'
mp4_codec = 'libx264'
mp4_bitrate = '5M'

video_clip = concatenate(clips, method="compose")
video_clip.write_videofile(mp4_file, fps=30, codec=mp4_codec, bitrate=mp4_bitrate)
```
